```python
import pandas as pd
import numpy as np
from glob import glob
from os import path

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ls ../ssrt_metrics
```

```python
METRIC_DIR = '../ssrt_metrics'

full_data = pd.DataFrame()
for sim_file in glob(path.join(METRIC_DIR, '*')):
    sim_key = sim_file.split('/')[-1].replace('.csv', '')
    curr_df = pd.read_csv(sim_file)
    curr_df['simulation_type'] = sim_key
    full_data = pd.concat([full_data, curr_df], 0)
    
```

```python
plt.figure(figsize=(20,12))
_ = sns.boxplot(x='simulation_type', y='SSRT_replacement', data=full_data)
```
