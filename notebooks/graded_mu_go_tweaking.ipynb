```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def graded_mu_go(SSD, max_SSD=550):
    return np.log(SSD/max_SSD)/4+1
```

```python
inputs = np.arange(0, 560, 10)
plt.figure(figsize = (10,10))
_ = plt.plot(inputs, graded_mu_go(inputs))
```

```python
graded_mu_go(550)
```
