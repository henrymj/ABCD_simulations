```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
pguess = pd.read_csv('p_guess_per_ssd.csv', index_col=0).query('SSD < 600')
plt.plot(pguess.SSD, pguess.p_guess)
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

def func(x, a, b, c):
    return a * x + b * x**2 + c
 

popt, pcov = curve_fit(func, pguess.SSD, pguess.p_guess)
```

```python
plt.plot(pguess.SSD, func(pguess.SSD, *popt))
plt.plot(pguess.SSD, pguess.p_guess, 'k')
```
