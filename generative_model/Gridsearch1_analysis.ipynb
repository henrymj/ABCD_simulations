```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sim_results = pd.read_csv('simulation_results2.csv', index_col=0)
```

```python
sim_results_abs = sim_results.copy()
sim_results_abs['goacc_error_abs'] = sim_results.goacc_error.abs()
sim_results_abs['stopfailrt_error_abs'] = sim_results.stopfailrt_error.abs()
sim_results_abs['gort_error_abs'] = sim_results.gort_error.abs()
sim_results_abs = sim_results_abs.drop(columns=['goacc_error', 'stopfailrt_error','gort_error' ])
```

```python
sns.pairplot(sim_results_abs)
```

```python

import statsmodels.api as sm

y = sim_results_abs['presp_rmse']
X = sim_results_abs[['mu_go', 'mu_stop_delta', 'mu_delta_incorrect']]
X = sm.add_constant(X)
mod = sm.OLS(y, X)
res = mod.fit()
print(res.summary())
```
